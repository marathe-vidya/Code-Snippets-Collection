```python
import pandas as pd
from pytrends.request import TrendReq
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
keyword_list = ['Twitter', 'Elon Musk']
time_interval = 'now 7-d'
pytrends.build_payload(keyword_list, cat=0, timeframe=time_interval, geo="ES")
```

```python
df = pytrends.interest_over_time()
print("total records fetached:",df.size)
df.head()
```

```python
from pytrends.request import TrendReq
import pandas as pd
pytrend = TrendReq(hl='en-US', tz=360)
pytrend.build_payload(kw_list=['NASDAQ:AAPL', 'NASDAQ:META'],timeframe = 'now 7-d')
df = pytrend.interest_over_time()
df2 = df.drop(['isPartial'], axis=1)
df2.to_csv('rndm.csv')
df2 = pd.read_csv('rndm.csv')
```

```python
newList = []
oldList = []
for item in df2['date']:
    oldList.append(item)
    newList.append(item.split()[0])
df2['date'] = df2['date'].replace(oldList, newList)
```
